```python
import matplotlib.pyplot as plt

#Morket share datu

manufacturers = ['Samsung', 'Apple', 'Huawei', 'Xiaomi', 'Others']

market_share= [30, 25, 18, 12, 15]

#Colars for each slice

colors = ['lightblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightgrey']

# Create a pie chart

plt.pie(market_share, labels=manufacturers, colors=colors, autopct='1.11')

#Title

plt.title('Smartphone Market Share')

# Show the pie chart

plt.show()
```

```python
import matplotlib.pyplot as plt 
labels = ['A', 'B', 'C', 'D']

sizes =[15, 30, 45, 10] 
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Creating pie chart with Legend 
plt.pie(sizes, labels= labels, colors=colors, autopct='%1.1f%%', startangle=140) 
plt.title('Pie Chart with Legend')

plt.legend(labels, loc="best")

plt.axis('equal') 
# Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import matplotlib.pyplot as plt

# Student data (study hours and exam scores)

study_hours = [2, 3, 1, 4, 3, 5, 2, 6, 5, 7]

exam_scores = [65, 75, 60, 80, 70, 85, 70, 90, 88, 92]

# Create a scatter plot

plt.scatter(study_hours, exam_scores, c='r', marker='o', label='Student Data')

# Labeling and Title

plt.xlabel('Study Hours')

plt.ylabel('Exam Scores')

plt.title('Study Hours vs. Exam Scores')

# Show the scatter plot

plt.legend()

plt.grid(True)

plt.show()
```
