```python
import matplotlib.pyplot as plt

#Sample student data

students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']

math_scores = [85, 92, 78, 88, 90] 
science_scores = [76, 88, 92, 80, 78]


# Create a subplot with two subplots (1 row, 2 columns)

fig, axs =plt.subplots (1, 2, figsize=(8, 4))

# Plot Math Scores

axs[0].bar(students, math_scores, color='b')

axs[0].set_title('Math Scores')

axs[0].set_xlabel('Students')

axs[0].set_ylabel('Score')

#Plot Science Scores

axs [1].bar(students, science_scores, color='g')

axs [1].set_title('Science Scores')

axs[1].set_xlabel('Students')

axs [1].set_ylabel('Score')

#Adjust spacing between subplots

plt.tight_layout()

# Show the subplot

plt.show()
```

```python
import matplotlib.pyplot as plt

#Sample data

x= [1, 2, 3, 4, 5]

y = [2, 3, 5, 7, 11]

#Creating plot

plt.plot(x, y)

#Adding title and axis Labels

plt.title('Plot with General Text')

plt.xlabel('X-axis')

plt.ylabel('Y-axis')

#Adding general text

plt.text(2, 8, 'Language', fontsize=10, color='white', bbox={"facecolor":"red"})

plt.annotate(r'Java', xy=(4, 7), xytext=(2, 10), arrowprops=dict(facecolor='yellow', shrink=0.10), fontsize=12, color='red')

plt.legend(["Java"], loc=5, facecolor="yellow", edgecolor="red", framealpha=0.9, shadow=True)

plt.show()
```

```python
# Matplotlib provides several functions to customize the appearance of the axes in a plot. These include setting the ticks and limits for both the 
# x-axis and y-axis. Here's how you can use xticks, yticks, xlim, and ylim in Matplotlib, along with some practical examples. 
# xticks and yticks: These functions are used to get or set the current tick locations and labels of the x-axis and y-axis, respectively. 
# xlim and ylim: These functions are used to get or set the current x-axis and y-axis limits, respectively.

x=[1,2,3,4,5,6,7,8]
y=[5,7,3,8,9,5,4,8]

plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,12)
plt.show()
```

```python
# xticks and yticks: These functions are used to get or set the current tick locations and labels of the x-axis and y-axis, respectively.

import matplotlib.pyplot as plt

# Sample data 
x = [1, 2, 3, 4, 5] 
y = [2, 3, 5, 7, 11]

#Creating plot 
plt.plot(x, y)

# Customizing ticks

plt.xticks([1, 2, 3, 4, 5], ['One', 'Two', 'Three', 'Four', 'Five'])

plt.yticks([2, 4, 6, 8, 10, 12], ['Two', 'Four', 'Six', 'Eight', 'Ten', 'Twelve'])

plt.title('Custom Ticks Example')

plt.xlabel('X-axis')

plt.ylabel('Y-axis')

#Showing plot

plt.show()
```
